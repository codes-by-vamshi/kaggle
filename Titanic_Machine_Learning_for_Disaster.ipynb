```python
import zipfile
zip_ref = zipfile.ZipFile('titanic.zip','r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
df = pd.read_csv('train.csv')
df.sample(5)
```

```python
df['coutesy_title'] = df['Name'].str.split(', ',expand=True)[1].str.split('. ',expand=True)[0]
df.sample(5)
```

```python
df.drop(columns=['Cabin','PassengerId'],inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Embarked'])
df.sample(5)
```

```python
df['coutesy_title'].unique()
```

```python
df.drop(columns=['Name','Ticket'],inplace=True)
df.sample()
```

```python
from sklearn.impute import SimpleImputer
sme = SimpleImputer()
df['Age'] = sme.fit_transform(df[['Age']])
```

```python
df.sample(5)
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
X = df.drop(['Survived','Parch'],axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
preds = np.round(model.predict(X_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(preds,y_test)
```

## With PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_train)

pca = PCA(n_components=1)
pca_result = pca.fit_transform(scaled_data)
```

```python
test_scaled_data = scaler.transform(X_test)
test_pca_result = pca.transform(test_scaled_data)
```

```python
model.fit(pca_result,y_train)
```

```python
preds = model.predict(test_pca_result)
```

```python
accuracy_score(preds,y_test)
```

```python
outpDf = pd.DataFrame()
outpDf['PassengerId'] = passId
outpDf['Survived'] = model.predict(pca.transform(scaler.transform(df_test)))
```

```python
outpDf = outpDf.set_index('PassengerId')
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_train)

pca = PCA(n_components=5)
pca_result = pca.fit_transform(scaled_data)
```

```python
test_scaled_data = scaler.transform(X_test)
test_pca_result = pca.transform(test_scaled_data)
```

```python
model = RandomForestClassifier(n_estimators=301, random_state=42)
```

```python
model.fit(pca_result,y_train)
```

```python
preds = model.predict(test_pca_result)
```

```python
accuracy_score(preds,y_test)
```

```python
outpDf = pd.DataFrame()
outpDf['PassengerId'] = passId
outpDf['Survived'] = model.predict(pca.transform(scaler.transform(df_test)))
```

```python
outpDf = outpDf.set_index('PassengerId')
```

```python
outpDf.to_csv('pred.csv')
```
